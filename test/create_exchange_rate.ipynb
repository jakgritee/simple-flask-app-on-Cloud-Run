```python
from datetime import datetime
import pandas as pd

date = pd.date_range(start=datetime(2010, 12, 1), end=datetime(2011, 12, 9))
date
```

```python
len(date)
```

```python
import numpy as np

rate = np.random.uniform(low=46, high=50, size=len(date))
rate
```
